```python
import pandas as pd

csv_files = [
    '/content/daegu1.csv',
    '/content/daegu2.csv',
    '/content/daegu3.csv',
    '/content/daegu4.csv',
    '/content/daegu5.csv',
    '/content/daegu6.csv',
    '/content/daegu7.csv',
    '/content/daegu8.csv'
]

filtered_data = []

for file in csv_files:
    try:
        df = pd.read_csv(file, encoding='cp949')
        if '도서관명' in df.columns:
            filtered = df[df['도서관명'].astype(str).str.contains('대구광역시립달성도서관', na=False)]
            if not filtered.empty:
                filtered_data.append(filtered)
    except Exception as e:
        print(f" 오류 발생: {file}, 내용: {e}")

if filtered_data:
    result_df = pd.concat(filtered_data, ignore_index=True)
    result_df.to_csv('/content/sample_data/daegu_lib.csv', index=False, encoding='utf-8-sig')
    print("/content/sample_data/daegu_lib.csv 저장됨")
else:
    print(" 해당 도서관 데이터가 없습니다.")
```

```python
import pandas as pd

# 원본 데이터 불러오기
df = pd.read_csv('/content/sample_data/daegu_lib.csv', thousands=',', encoding='utf-8')

# 날짜 형식 변환
df['대출일자'] = pd.to_datetime(df['대출일자'], errors='coerce')

# 도서관명
target_name = '대구광역시립달성도서관'

# 2023년 데이터
df_2023 = df[
    (df['도서관명'] == target_name) &
    (df['대출일자'].dt.year == 2023)
]

# 2024년 데이터
df_2024 = df[
    (df['도서관명'] == target_name) &
    (df['대출일자'].dt.year == 2024)
]

# 두 연도 합치기
df_combined = pd.concat([df_2023, df_2024], ignore_index=True)

# 결과 확인
print(f"2023년: {len(df_2023)}건, 2024년: {len(df_2024)}건, 총합: {len(df_combined)}건")
print(df_combined[['성별', '서명', '대출일자', '도서관명']].head())

# 저장 (선택)
df_combined.to_csv('daegu_lib2324.csv', index=False, encoding='utf-8-sig')
print("daegu_lib2324.csv")
```
